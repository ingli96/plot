```python
x = np.arange(0, 10, 0.1) # on x axis we will use 100 values [0,0.1,0.2...9.9]
y = np.sin(x) # on y axis we will calculate sin for each value x
```

```python
plt.plot(x, y,   color='blue')        # color name
plt.plot(x, y-1, color='g')           # short color code (rgbcmyk)
plt.plot(x, y-2, color='0.5')         # Grayscale between 0 and 1
plt.plot(x, y-3, color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, y-4, color='DARKORCHID')  # HTML color name
plt.plot(x, y-5, color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.show()
```

```python
plt.plot(x, y - 1, ls='-')  # solid lne
plt.plot(x, y - 2, ls='--') # dashed line
plt.plot(x, y - 3, linestyle='-.') # dash-dotted line
plt.plot(x, y - 4, linestyle=':')  # dotted line
plt.show()
```

```python
plt.plot(x, y - 1, lw = 1)  # default width
plt.plot(x, y - 2, lw= 3) 
plt.plot(x, y - 3, linewidth = 5) 
plt.plot(x, y - 4, linewidth =0.5) # we can also make line thinner
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,np.cos(x))
plt.xlim(-1, 11) # we change range of x from (0,10)
plt.ylim(-1.5, 1.5) # we change range of x from (-1,1)
plt.show()
```

```python
plt.plot(x, y, label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.title("A Sine and Cos Curves", loc= 'left', pad = 20,color = '#0c4a91')
plt.xlabel("x", labelpad = 10, size = 'large', color = '#3383de' )
plt.ylabel("rad", labelpad = 20, size = 'large', color ='#3383de' )
plt.legend( loc='lower center', ncol=2, shadow = True, fontsize = 15)
# plt.axis('scaled')
plt.axis('equal')
plt.show()
```

```python
plt.plot(x,y)
plt.text(4,1,'Some text', size=12, color='#b50b1c') # we added cusotm text on position (4,1)
plt.annotate('minimum', xy=(x[np.argmin(y)],y.min()), xytext=(8,-0.5), color = '#b50b1c', arrowprops={'arrowstyle':"-|>"} )
plt.show()
# plt.figure(figsize=(w,h), dpi, facecolor...) 
```

```python
x = np.arange(0.1, 2, 0.1)
y = np.log(x)
y1 = np.log2(x)
y2 = np.log10(x)
```

```python
# plt.figure(figsize=(10,10), dpi=200, facecolor='#000000') 
plt.plot(x, y, ls=':', label='log')
plt.plot(x, y1, ls="-.", label='log2')
plt.plot(x, y2, ls='--', label='log10')
plt.annotate('intersection', xy=(1,0), xytext=(1.5,-0.5), color = '#b50b1c', arrowprops={'arrowstyle':"-|>"})
plt.title("Text", loc= 'right', pad = 10,color = '#000000')
plt.legend( loc='lower center', ncol=2, shadow = True, fontsize = 10)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)

y = np.sin(x)
```

```python
fig, axes = plt.subplots(2) # we store returned Figure and Axes in variables fig and axes
axes[0].plot(x, y) # we are plotting on first Axe from array axes
plt.show()
```

```python
fig, ax = plt.subplots(2,2) # we are creating subplots with 2 rows and 2 columns
ax[0,0].plot(x, np.sin(x)) # Axes in first column and first row
ax[0,0].set(title = 'Sin')
ax[0,1].plot(x, np.cos(x)) # Axes in first column and second row...
ax[0,1].set(title = 'Cos')
ax[1,0].plot(x, np.sqrt(x))
ax[1,0].set(title = 'Sqrt')
ax[1,1].plot(x, np.tan(x))
ax[1,1].set(title = 'Tan')
plt.show()
```

```python
fig, ax = plt.subplots(2,2) 

ax[0,0].plot(x, np.sin(x), c = '#e6a40b' ) 
ax[0,0].set_title('Sin', c = '#e6a40b', size = 20)

ax[0,1].plot(x, np.cos(x), c = '#e6a40b')
ax[0,1].set(title = 'Cos')

ax[1,0].plot(x, np.sqrt(x), c = '#e6a40b')
ax[1,0].set(title = 'Sqrt')

ax[1,1].plot(x, np.tan(x), c = '#e6a40b')
ax[1,1].set(title = 'Tan')

fig.set_size_inches(15, 5)
fig.set_facecolor('#9fcbf5')
fig.set_dpi(200)
plt.show()
```

```python
# data we will be working with
states = ['California', 'Texas', 'New York', 'Florida', 'Illinois','Pennsylvania','Ohio', 'New Jersey', 'Georgia', 'Washington']
gdp_2018 = [3.02, 1.82, 1.7, 1.06, 0.88, 0.8, 0.69, 0.63, 0.6, 0.58]
gdp_2017 = [2.75, 1.7, 1.55, 0.87, 0.82, 0.75, 0.65, 0.6, 0.65, 0.51 ]

indx = np.arange(len(states)) # we will assign number to each state (0,1,2,3...)
width = 0.4 # this will be a width of a column (distance between states is 1 so width of our column needs to be smaller than 0.5,
# since we will need 2 columns on each index)
plt.figure(figsize=(10,5), dpi = 150) 
plt.bar(indx - width/2, gdp_2018, width, color = '#d6132d', edgecolor = 'k', label = 'GDP 2018') 
plt.bar(indx + width/2, gdp_2017, width, color = '#2238a8', edgecolor = 'k', label = 'GDP 2017') 
plt.title ('US states by GDP ')
plt.legend()
plt.xticks(indx, states) # we will add names again for each tick on x axis
plt.show()
```

```python
plt.scatter(random_x ,random_y,c=colors, s=sizes, alpha=0.3, cmap= 'inferno') # we will add colormap
plt.colorbar() # this will show colour scale 
plt.show()
```

```python
# tickers
import matplotlib.ticker as ticker

x = range(1,101)
y = np.random.randint(1, 10, 100)

fig, ax = plt.subplots()
fig.set_size_inches(15,5)
ax.plot(x, y)

ax.xaxis.set_major_locator(ticker.MultipleLocator(20))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(10))

ax.yaxis.set_major_locator(ticker.MultipleLocator(3))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(1))
```
